## Types of statistical distributions

Depending on the type of data we use, we have grouped distributions into two categories, discrete distributions for discrete data (finite outcomes) and continuous distributions for continuous data (infinite outcomes).

## 1. Normal Distribution:

### Introduction:

The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that plays a fundamental role in statistics and data analysis. It is characterized by a symmetric, bell-shaped curve, where data is centered around the mean and gradually decreases as values move away from the mean in either direction. The normal distribution is fully defined by its two parameters: the mean (μ) and the standard deviation (σ).

### When to Use:

The normal distribution is widely used in various fields and scenarios due to its versatility and mathematical properties. It is appropriate to use the normal distribution under the following conditions:

- When dealing with a continuous random variable where values can theoretically range from negative infinity to positive infinity.

- When the data is reasonably symmetric and unimodal (one peak) in nature.

- When the data is not significantly skewed and does not exhibit heavy tails (outliers).

- When applying parametric statistical tests and modeling techniques that assume normality, such as t-tests, ANOVA, linear regression, and many more.

### Advantages:

- Widely Applicable: The normal distribution is used in a wide range of statistical applications and is a good approximation for many natural phenomena.

- Simple and Well-Defined: The normal distribution is fully defined by just two parameters, making it mathematically tractable and easy to work with.

- Central Limit Theorem: The normal distribution arises as a result of the Central Limit Theorem, which states that the sum of a large number of independent and identically distributed random variables tends to follow a normal distribution, even if the individual variables do not.

### Disadvantages:

- Sensitivity to Assumptions: The normal distribution assumes that the data is continuous and normally distributed. If this assumption is violated, it can lead to misleading or incorrect results when using normality-based statistical tests and models.

- Limited Applicability to Skewed Data: The normal distribution is not suitable for highly skewed data or data with heavy tails. In such cases, non-parametric approaches or other distributions may be more appropriate.

- Outliers Can Distort Results: The normal distribution is sensitive to outliers, which can affect estimates of the mean and standard deviation, leading to biased results.

###  Alternatives:

When the normal distribution assumptions are not met or when dealing with non-normal data, alternative distributions may be considered. Some common alternatives include:

- Log-Normal Distribution: Appropriate for positively skewed data where the logarithm of the data follows a normal distribution.

- Exponential Distribution: Suitable for modeling time between events in processes with constant failure rates.

- Weibull Distribution: Used for modeling failure rates that change over time, allowing for different shapes of hazard functions.

- Binomial Distribution: Describes the number of successes in a fixed number of independent Bernoulli trials.

- Poisson Distribution: Used to model the number of events occurring in fixed intervals of time or space.

### Types of Normal Distribution:

There are different types of normal distributions based on the values of the mean and standard deviation:

- Standard Normal Distribution: A special case of the normal distribution with a mean (μ) of 0 and a standard deviation (σ) of 1.

- Standardized Normal Distribution: Any normal distribution can be transformed into a standard normal distribution by the process of standardization.

#### Formula:

The probability density function (PDF) of the normal distribution is given by the formula:

f(x) = (1 / (σ * √(2π))) * exp(-((x - μ)² / (2σ²)))

where:

- x is the random variable (a data point)

- μ is the mean of the distribution

- σ is the standard deviation of the distribution

- π (pi) is a mathematical constant approximately equal to 3.14159

- exp() represents the exponential function

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 50  # Mean of the distribution
std_dev = 10  # Standard deviation of the distribution
num_samples = 1000  # Number of data points to generate

# Generate random data following a normal distribution
data = np.random.normal(mean, std_dev, num_samples)

# Plot the histogram to visualize the distribution
plt.hist(data, bins=20, density=True, alpha=0.75, edgecolor='black')
plt.xlabel('Data Points')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.grid(True)

# Plot the theoretical probability density function (PDF) of the normal distribution
x = np.linspace(min(data), max(data), 1000)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))
plt.plot(x, pdf, color='red', linewidth=2, label='Theoretical PDF')

plt.legend()
plt.show()
```

## Skewed Distribution:

A distribution is considered skewed when it is not symmetrical. Skewness can be either positively skewed (tail extends towards higher values) or negatively skewed (tail extends towards lower values). Skewed distributions indicate that data is not evenly distributed and may be influenced by outliers.

Introduction:
A skewed distribution is a type of probability distribution where the data is not symmetrically distributed around the mean and exhibits a tail towards one direction. In a skewed distribution, the data points cluster more in one tail of the distribution than the other. Skewness is a measure of the asymmetry in the distribution.

When to Use:
Skewed distributions commonly arise in various real-world scenarios, especially when dealing with non-negative data or financial data. Skewness can have important implications for data analysis, as it may affect the choice of statistical tests and models. Understanding the skewness of data is essential when making decisions and drawing insights from the data.

Advantages:

Identifying Outliers: Skewed distributions can help identify potential outliers or extreme values in the dataset.
Insights into Data: Skewness provides insights into the underlying patterns and characteristics of the data, indicating where most of the data points are concentrated.
Disadvantages:

Impact on Assumptions: Skewness can violate the assumptions of normality in statistical tests and models, leading to incorrect conclusions if not properly addressed.
Limited Applicability of Parametric Tests: Parametric tests that assume normality may not be suitable for skewed data, necessitating the use of non-parametric alternatives.
Alternatives:
When dealing with skewed data, alternatives to normality-based analysis may include:

Non-parametric Tests: These statistical tests do not assume a specific distribution, making them robust to skewed data.
Data Transformation: Applying data transformations, such as logarithmic or power transformations, can help reduce skewness and make the data more suitable for parametric analysis.
Types of Skewed Distributions:

Positive Skew (Right Skew): The tail extends towards higher values, and the majority of data points are concentrated on the left side of the distribution.
Negative Skew (Left Skew): The tail extends towards lower values, and most data points are concentrated on the right side of the distribution.
Formula:
The skewness of a dataset is typically quantified using the Pearson's skewness coefficient, defined as:

Skewness = (3 * (Mean - Median)) / Standard Deviation

If the skewness value is:

Positive (> 0): The distribution is positively skewed.
Negative (< 0): The distribution is negatively skewed.
Close to 0: The distribution is approximately symmetric.
Plot Graph and Explanation:

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the skewed distribution
mean = 50
std_dev = 10
num_samples = 1000

# Generate positively skewed data
data_pos_skewed = np.random.gamma(2, 1, num_samples) * std_dev + mean

# Calculate skewness
skewness = (3 * (np.mean(data_pos_skewed) - np.median(data_pos_skewed))) / np.std(data_pos_skewed)

# Plot the histogram
plt.hist(data_pos_skewed, bins=20, density=True, alpha=0.75, edgecolor='black')
plt.xlabel('Data Points')
plt.ylabel('Probability Density')
plt.title('Positively Skewed Distribution')
plt.grid(True)

plt.show()

print("Skewness:", skewness)
```

## Uniform Distribution:

In a uniform distribution, all values have an equal probability of occurring. It is a rectangular-shaped distribution, where the data points are spread uniformly across the range. Examples of uniform distributions can be found in random number generators or in scenarios where each outcome is equally likely.

## Bimodal Distribution:

A bimodal distribution exhibits two distinct peaks, indicating that the data has two modes (frequent values). This could happen due to the presence of two different groups or subpopulations within the data.

## Exponential Distribution:

The exponential distribution is often used to model the time between events in a process that occurs continuously and independently at a constant rate. It is characterized by a rapid drop-off in frequency as values increase.

### Creating the skewness function

```python
def skew(data,col):
    sns.kdeplot(data=data, x= col)
    plt.show()
  
    mean = round(data[col].mean(),2)
    median = round(data[col].median(),2)
    mode = round(3*median - 2*mean,2)
    skew = round(data[col].skew(),2)
    if mean == median:
        print("Mean == Median == Mode \nData is Symmentric Data\n")
    else:
        print("Mean != Median != Mode \nData is Asymmentric Data\n")
        if mean < median:
            print("\nLeft or Negative skeweness \nSkewness Value:",skew)
        else:
            print("\nRight or Positive skeweness \nSkewness Value:",skew)
    
    print("\nMean of",col,':',mean)
    print("Median of",col,':',median)
    print("mode of",col,':',mode)
```

```python
def kurtosis(data,col):
    sns.kdeplot(data=data, x= col)
    plt.show()
  
    mean = round(data[col].mean(),2)
    median = round(data[col].median(),2)
    mode = round(3*median - 2*mean,2)
    skew = round(data[col].skew(),2)
    kur = round(data[col].kurt(),2)
    if mean == median:
        print("Mean == Median == Mode \nData is Symmentric Data\n")
        if kur < 3:
            print("Kurtosis value",kur,"\nData is having platy or Negative or flat kurtosis")
        elif kur == 3:
            print("Kurtosis value",kur,"\nData is having Meso or Equal or No Kurtosis")
        else:
            print("Kurtosis value",kur,"\nData is having Lepto or Positive or thin kurtosis")
    else:
        print("Mean != Median != Mode \nData is Asymmentric Data\n")
        if mean < median:
            print("\nLeft or Negative skeweness \nSkewness Value:",skew)
        else:
            print("\nRight or Positive skeweness \nSkewness Value:",skew)
    
    print("\nMean of",col,':',mean)
    print("Median of",col,':',median)
    print("mode of",col,':',mode)
```
