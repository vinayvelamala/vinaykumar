### Plotting x and y points

The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 6])
y = np.array([3, 8, 1, 10])

plt.plot(x, y)
plt.show()
```

### Create Labels for a Plot

```python
plt.plot(x,y,'o:b')

plt.xlabel("Speed")
plt.ylabel("Time")

plt.show()
```

### Create a Title for a Plot

```python
plt.plot(x,y,'o:b')

plt.title("Sports Watch Data")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

### Add Grid Lines to a Plot

```python
plt.plot(x,y,'o:b')

plt.title("Sports Watch Data")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.grid()

plt.show()
```

```python
plt.plot(x,y,'o:b')

plt.title("Sports Watch Data")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

### Display Multiple Plots

#### plt.subplot(1, 2, 1)

the figure has 1 row, 2 columns, and this plot is the first plot.

#### plt.subplot(1, 2, 2)

the figure has 1 row, 2 columns, and this plot is the second plot.

```python
#plot 1:
plt.subplot(1, 3, 1)
plt.plot(x,y,'o:b')

#plot 2:
plt.subplot(1, 3, 3)
plt.plot(x,y,'o:r')

#plot 3:
plt.subplot(1, 3, 2)
plt.plot(x,y,'o:y')

plt.show()
```

```python
#plot 1:
plt.subplot(2, 1, 1)
plt.plot(x,y,'o:b')

#plot 2:
plt.subplot(2, 1, 2)
plt.plot(x,y,'o:r')

plt.show()
```

```python
# plot 1
plt.subplot(2, 3, 1)
plt.plot(x,y,'s:b')

# plot 2
plt.subplot(2, 3, 2)
plt.plot(x,y,'s:r')

# plot 3
plt.subplot(2, 3, 3)
plt.plot(x,y,'s:m')

# plot 4
plt.subplot(2, 3, 4)
plt.plot(x,y,'o:r')

# plot 5
plt.subplot(2, 3, 5)
plt.plot(x,y,'*:r')

# plot 6
plt.subplot(2, 3, 6)
plt.plot(x,y,'X:r')

plt.show()
```

### Pie Chart

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0,0,0,0.2]
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```
